# Creating a Wordcloud

### Importing new libraries needed

```python
import numpy as np
from os import path
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

### Importing the text

```python
d = path.dirname(r'C:\Users\Luca\OneDrive\Documenti\University\Master in Data Science\Data Storytelling\text.txt')

text = open(path.join(d, 'text.txt')).read()
```

### Setting stopwords to clean the text

```python
stopwords = set(STOPWORDS)
stopwords.add("light")
stopwords.add("lights")
stopwords.add("NUFORC")
stopwords.add("Note")
stopwords.add("UFO")
stopwords.add("across")
stopwords.add("sky")
stopwords.add("saw")
stopwords.add("one")
stopwords.add("two")
stopwords.add("night")
```

### Defining the Wordcloud Function

```python
wc = WordCloud(width=1600, height=800,max_words=1000, stopwords=stopwords, margin=1,
               random_state=1,prefer_horizontal=0.3,background_color="white").generate(text)
```

### Plotting the Wordcloud

```python
plt.figure(figsize=(20,20))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
