# Creating a Wordcloud

### Importing new libraries needed

```python
from os import path
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

### Importing the text and the mask

```python
d = path.dirname(r'C:\Users\Luca\OneDrive\Documenti\University\Master in Data Science\Data Storytelling\text.txt')
text = open(path.join(d, 'text.txt')).read()
mask = np.array(Image.open(path.join(d, "mask.png")))
```

### Setting stopwords to clean the text

```python
stopwords = set(STOPWORDS)
stopwords.add("light")
stopwords.add("lights")
stopwords.add("NUFORC")
stopwords.add("Note")
stopwords.add("UFO")
stopwords.add("across")
stopwords.add("sky")
stopwords.add("saw")
stopwords.add("one")
stopwords.add("two")
stopwords.add("night")
```

### Defining the Wordcloud Function

```python
wc = WordCloud(max_words=1000, stopwords=stopwords, margin=1,
               random_state=1,prefer_horizontal=0.3,background_color="white", mask=mask)
```

### Plotting the Wordcloud

```python
wc.generate(text)
wc.to_file(path.join(d, "wordcloud_mask.png"))
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
